## Q1 Predict delivery time using sorting **time**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
```

```python
data=pd.read_csv('/content/delivery_time.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
plt.boxplot(data['Delivery Time'])
```

```python
plt.boxplot(data['Sorting Time'])
```

```python
sns.distplot(data['Delivery Time'])
```

```python
sns.distplot(data['Sorting Time'])
```

```python
sns.pairplot(data)
```

```python
data=data.rename({'Delivery Time':'delivery_time', 'Sorting Time':'sorting_time'},axis=1)
data                          #Rename function
```

```python
data.corr()           #Correlation Analysis
```

```python
sns.regplot(x=data['sorting_time'],y=data['delivery_time'])
```

```python
import statsmodels.formula.api as smf
model1 = smf.ols("delivery_time~sorting_time",data = data).fit()
```

```python
print(model1.params,'\n''\n',model1.tvalues,'\n''\n', model1.pvalues)
```

```python
#R squared values
(model1.rsquared,model1.rsquared_adj)
```

```python
model1.summary()
```

```python
model2 = smf.ols("np.log(delivery_time)~np.log(sorting_time)",data = data).fit()
```

```python
print(model2.params,'\n''\n',model2.tvalues,'\n''\n', model2.pvalues)
```

```python
#R squared values
(model2.rsquared,model2.rsquared_adj)
```

```python
model2.summary()
```

```python
model3 = smf.ols("delivery_time~np.square(sorting_time)",data = data).fit()
```

```python
print(model3.params,'\n''\n',model3.tvalues,'\n''\n', model3.pvalues)
```

```python
#R squared values
(model3.rsquared,model3.rsquared_adj)
```

```python
model3.summary()
```

```python
model4 = smf.ols("delivery_time~np.sqrt(sorting_time)",data = data).fit()
```

```python
print(model4.params,'\n''\n',model4.tvalues,'\n''\n', model4.pvalues)
```

```python
#R squared values
(model4.rsquared,model4.rsquared_adj)
```

```python
model4.summary()
```

## Q2 Build a prediction model for **Salary_hike**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
import statsmodels.formula.api as smf
```

```python
salary = pd.read_csv('/content/Salary_Data.csv')
salary
```

```python
salary.info()
```

```python
salary.describe()
```

```python
salary.isnull().sum()
```

```python
salary.shape
```

```python
sns.distplot(salary['YearsExperience'])
```

```python
sns.distplot(salary['Salary'])
```

```python
sns.pairplot(salary)
```

```python
salary.corr()          #Correlation Analysis
```

```python
salary = salary.rename({'YearsExperience':'Experience'},axis=1)
salary                          #Rename function
```

```python
sns.regplot(x=salary['Experience'],y=salary['Salary'])
```

```python
model1 = smf.ols("Salary~Experience",data = salary).fit()
```

```python
print(model1.params,'\n''\n',model1.tvalues,'\n''\n', model1.pvalues)
```

```python
#R squared values
(model1.rsquared,model1.rsquared_adj)
```

```python
model1.summary()
```

```python
#Prediction for Salary hike
newdata=pd.Series([7,15])
data_pred=pd.DataFrame(newdata,columns=['Experience'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
model2 = smf.ols("Salary~np.log(Experience)",data = salary).fit()   #using log
```

```python
print(model2.params,'\n''\n',model2.tvalues,'\n''\n', model2.pvalues)
```

```python
#R squared values
(model2.rsquared,model2.rsquared_adj)
```

```python
model2.summary()
```

```python
model3 = smf.ols("Salary~np.square(Experience)",data = salary).fit()   #using square
```

```python
print(model3.params,'\n''\n',model3.tvalues,'\n''\n', model3.pvalues)
```

```python
#R squared values
(model3.rsquared,model3.rsquared_adj)
```

```python
model3.summary()
```

```python
model4 = smf.ols("Salary~np.sqrt(Experience)",data = salary).fit()   #using square
```

```python
print(model4.params,'\n''\n',model4.tvalues,'\n''\n', model4.pvalues)
```

```python
#R squared values
(model4.rsquared,model4.rsquared_adj)
```

```python
model4.summary()
```
